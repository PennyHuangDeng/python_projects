```python
plt.figure(figsize=(6,4),dpi=100)
ax = plt.subplot(111)

sizes = [22951,9853,18]
labels = ['男','女','未知'] 
colors = ['#4169E1','#FFA500','#FF0000']
explode = [0, 0, 0]
patches,l_text = plt.pie(sizes,
                       colors=colors, 
                       wedgeprops = dict(width = 0.3),
                       startangle = 0,
                       textprops = dict(fontsize = 15),
                       )
ax.set_facecolor('none')

plt.text(-0.5, 0.2, '69.93%', fontsize = 30,color = '#4169E1',family = 'Songti SC')
plt.text(-0.2, -0.4, '30.02%', fontsize = 24,color = '#FF8C00',family = 'Songti SC')
plt.text(-0.1, -0.1, '/', fontsize = 28,color = 'grey',family = 'Songti SC')


plt.text(-1.6, 1.5, '早起Python粉丝性别分布', fontsize = 16,color = '#0000CD',family = 'Songti SC')

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")


leg = plt.legend(patches, labels, loc=0, ncol = 3,fontsize = 7, bbox_to_anchor=(0.75, -0.05),handlelength=1, handleheight=1,prop = 'Songti SC',frameon=False)

frame = leg.get_frame() 
frame.set_alpha(1) 
frame.set_facecolor('none') # 设置图例legend背景透明 

for text in leg.get_texts():
    plt.setp(text, color = 'grey')
    
plt.axis('equal')

plt.show()
```

```python
import seaborn as sns

sns.set_palette("Blues", 8)

fig, ax = plt.subplots(figsize=(10, 5), subplot_kw=dict(aspect="equal"),dpi = 100)

plt.rcParams['font.sans-serif'] = ['SongTi SC']

recipe = ["北京：3897人",
          "上海：2973人",
          "广州：2032人",
          "深圳：1798人",
          "杭州：1084人"]
data = [3837, 2973, 2032, 1798, 1084]
wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5,linewidth= 1,edgecolor='black'), startangle=-40)
bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

    
plt.show()
```
