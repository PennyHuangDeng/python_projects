# Matplotlib图鉴——高级饼图

## 公众号：可视化图鉴

```python
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib.patches import Wedge
from matplotlib.offsetbox import TextArea, DrawingArea, OffsetImage, AnnotationBbox
%config InlineBackend.figure_format = 'retina'
plt.rcParams['font.sans-serif'] = ['Songti SC']
```

```python


plt.figure(figsize=(10,7),dpi = 120) 
data = [2806/4785/2,1335/4785/2,524/4785/2,72/4785/2,48/4785/2]

series = pd.Series(data)

label = ['5星力荐', '4星推荐','3星还行','2星较差','1星很差']
colors = ['#6C7B8B','#8DB6CD','#A4D3EE','#B0E2FF','#BFEFFF'][::-1] 
patches,l_text = plt.pie(series,
                      colors = colors,
                      wedgeprops = dict(width = 0.4),
                      startangle = 0,
                      textprops = dict(fontsize = 15),
                      normalize=False
                     )

plt.text(-0.9, 0.15, '11.0%', fontsize = 18,rotation = 345)
plt.text(-0.6, 0.52, '27.9%', fontsize = 18,rotation = 300)
plt.text(0.49, 0.45, '58.6%', fontsize = 18,rotation = 38)

ax = plt.gca()

semicircle=Wedge((0,0),1.05,0,180, edgecolor = 'grey', facecolor = 'None', linestyle='--')
ax.add_artist(semicircle)
plt.axis('equal')

semicircle=Wedge((0,0),0.55,0,180, edgecolor = 'grey', facecolor = 'None', linestyle='-')
ax.add_artist(semicircle)
plt.axis('equal')

x = [-1.05,1.05]
y = [0,0]
plt.plot(x, y,color='white',alpha=1)
plt.plot(x, y,color='white',alpha=1)
plt.plot(x, y,color='white',alpha=1)
plt.plot(x, y,color='white',alpha=1)
plt.plot(x, y,color='white',alpha=1)
plt.plot(x, y,color='white',alpha=1)
plt.plot(x, y,color='white',alpha=1)
plt.plot(x, y,color='white',alpha=1)
plt.plot(x, y,color='white',alpha=1)
plt.plot(x, y,color='white',alpha=1)

plt.text(-0.49, 0.22, '《阿凡达》星级分布', family = 'Songti SC', fontsize = 25)

plt.text(-0.5, 0.05, '数据来源：豆瓣电影  采集时间：4/5/2021', family = 'Songti SC', fontsize = 13,alpha = 0.7)

plt.legend(patches, label, loc=0, ncol = 5, fontsize = 13, bbox_to_anchor=(0.92, 0.15),handlelength=1, handleheight=1)


offsetbox = TextArea("差评 2.5%",textprops = dict(fontsize = 10))
ab = AnnotationBbox(offsetbox, [-1,0.05],
                     xybox=(-30, 30),
                    xycoords='data',
                    boxcoords="offset points",
                    arrowprops=dict(arrowstyle="->",connectionstyle="angle3,angleA=90,angleB=0"))
                    
                
ax.add_artist(ab)


plt.show()
```
