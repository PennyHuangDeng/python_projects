### 基础散点图—分组散点图

```python
x = list(np.arange(0,5,0.25))
y1 = np.random.rand(20)
y2 = np.random.rand(20)
plt.figure(figsize=(10,7))#设置画布的尺寸

plt.scatter(x,y1,s =150,marker ='^',alpha = 0.7)
plt.scatter(x,y2,s =150,marker ='o',alpha = 0.7)

plt.title("基础散点图—分组散点图",fontsize =20) # 图形标题
plt.xlabel("我是x轴",fontsize = 14) # x轴名称
plt.ylabel("我是y轴",fontsize = 14) # y轴名称

plt.show()
```

### 基础散点图—分组散点图—添加图例、颜色条

```python
x = list(np.arange(0,5,0.25))
y1 = np.random.rand(20)
y2 = np.random.rand(20)
plt.figure(figsize=(10,7))#设置画布的尺寸

plt.scatter(x,y1,s =150,marker ='^',alpha = 0.7,c = np.random.rand(20),label = "系列1",cmap=plt.get_cmap('jet'))
plt.scatter(x,y2,s =150,marker ='o',alpha = 0.7,c = np.random.rand(20),label = "系列2",cmap=plt.get_cmap('jet'))

plt.title("基础散点图—分组散点图—添加图例、颜色条",fontsize =16) # 图形标题
plt.xlabel("我是x轴",fontsize = 14) # x轴名称
plt.ylabel("我是y轴",fontsize = 14) # y轴名称

plt.legend(loc = 1,fontsize = 14)
plt.colorbar()
plt.show()
```

### 基础散点图—分组散点图—多组数据

```python
from sklearn import datasets 
iris=datasets.load_iris()
x, y = iris.data, iris.target
pd_iris = pd.DataFrame(np.hstack((x, y.reshape(150, 1))),columns=['sepal length(cm)','sepal width(cm)','petal length(cm)','petal width(cm)','class'] )
 
plt.figure(figsize = (10,7),dpi = 100)

plt.title("基础散点图—分组散点图—多组数据",fontsize =16) # 图形标题
plt.xlabel("我是x轴",fontsize = 12) # x轴名称
plt.ylabel("我是y轴",fontsize = 12) # y轴名称

iris_type=pd_iris['class'].unique()#将点分为三类
iris_name=iris.target_names
colors = ['#FF69B4','#098154','#fb832d']
markers = ['x','o','+'] #设置样式
 
for i in range(len(iris_type)):
    plt.scatter(pd_iris.loc[pd_iris['class'] == iris_type[i], 'sepal length(cm)'],
                pd_iris.loc[pd_iris['class'] == iris_type[i], 'sepal width(cm)'],
                s = 50,
                c = colors[i],
                marker = markers[i],#marker形状
                label = iris_name[i])#后面图例的名称取自label

plt.legend(loc = 1,fontsize = 13)
plt.show()
```
