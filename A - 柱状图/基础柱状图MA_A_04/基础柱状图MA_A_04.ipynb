### 基础柱状图——添加填充

```python
x = ['周一', '周二', '周三', '周四', '周五', '周六', '周日']
y = [58, 120, 86, 20, 139, 57, 139]
plt.figure(figsize=(10,7))#设置画布的尺寸

plt.bar(x = x,height = y, width = 0.7,alpha = 0.5,color = 'deeppink',edgecolor = 'black', linewidth = 2, ecolor = 'blue',label='我是图例',hatch = '/')

plt.title('基础柱状图—添加填充样式',fontsize=18) #设置标题
plt.xlabel('我是X轴',fontsize=14) #设置x轴
plt.ylabel('我是Y轴',fontsize=14) #设置y轴

plt.show()
```

### 基础柱状图——填充不同样式

```python
import random

x = ['周一', '周二', '周三', '周四', '周五', '周六', '周日']
y = [58, 120, 86, 20, 139, 57, 139]
h_list = ['/','o','-','+','x','.','*']
plt.figure(figsize=(10,7))#设置画布的尺寸

for i in range(len(x)):
        
    plt.bar(x = x[i],height = y[i], width = 0.7,alpha = 0.5,color = 'deeppink',edgecolor = 'black', linewidth = 2, ecolor = 'blue',label='我是图例',hatch = h_list[i])

plt.title('基础柱状图—填充不同样式',fontsize=18) #设置标题
plt.xlabel('我是X轴',fontsize=14) #设置x轴
plt.ylabel('我是Y轴',fontsize=14) #设置y轴

plt.show()
```

### 基础柱状图——堆叠柱状图——填充

```python
x = ['周一', '周二', '周三', '周四', '周五', '周六', '周日']
y1 = [7,6,5,4,3,2,1]
y2 = [1,2,3,3,1,1,0.5]

plt.figure(figsize=(10,7))#设置画布的尺寸

plt.bar(x, y1, label="系列1",edgecolor = 'black',hatch = '/')
plt.bar(x, y2, label="系列2",edgecolor = 'black',hatch = 'x')

plt.legend(loc=1,fontsize=14)  # 设置图例位置
plt.ylabel('我是Y轴',fontsize=16)
plt.xlabel('我是X轴',fontsize=16)
plt.title("基础柱状图——堆叠柱状图——填充",fontsize=18)

plt.show()
```

### 基础柱状图——分组柱状图—填充

```python
x = np.array(list(range(7)))
y1 = [7,6,5,4,3,2,1]
y2 = [6,2,3,3,1,1,0.5]

width = 0.4
x = x - 0.2

plt.figure(figsize=(9,6))

plt.bar(x, y1,  width=width, label='系列1',edgecolor = 'black',alpha = 0.7, hatch="/")
plt.bar(x + width, y2, width=width, label='系列2',edgecolor = 'black',hatch = 'x')

plt.xticks(x + 0.2,['周一', '周二', '周三', '周四', '周五', '周六', '周日'])


plt.legend(loc=1,fontsize=13)  # 设置图例位置
plt.ylabel('我是Y轴',fontsize=15)
plt.xlabel('我是X轴',fontsize=15)
plt.title("基础柱状图——分组柱状图——填充",fontsize=17)

plt.show()
```
