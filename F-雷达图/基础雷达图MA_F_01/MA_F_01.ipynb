### 基础雷达图

```python
import matplotlib.pyplot as plt 
import numpy as np 
    
types = ["A", "B", "C", 
            "D", "E"] 
  
actual = [41, 57, 27, 53, 52, 41] 
    
plt.figure(figsize =(9, 8)) 
plt.subplot(polar = True) 
    
theta = np.linspace(0, 2 * np.pi, len(actual)) 
    
lines, labels = plt.thetagrids(range(0, 360, int(360/len(types))), 
                                                         (types)) 
plt.plot(theta, actual) 

plt.legend(labels =('蓝'),bbox_to_anchor=(1, 0.2, 0.2, 0.8), fontsize=15) 

plt.title('我是标题', fontsize = 20)
plt.tick_params(labelsize=20)
plt.savefig('G_01_01.png')
    
plt.show()
```

### 基础雷达图 - 两组数据对比

```python
import matplotlib.pyplot as plt 
import numpy as np 
    
types = ["A", "B", "C", 
            "D", "E"] 
  
actual = [41, 57, 34, 53, 60, 41] 
expected = [40, 49, 20, 54, 65, 40] 
    
plt.figure(figsize =(9, 8)) 
plt.subplot(polar = True) 
    
theta = np.linspace(0, 2 * np.pi, len(actual)) 
    
lines, labels = plt.thetagrids(range(0, 360, int(360/len(types))), 
                                                         (types)) 
plt.plot(theta, actual, color = 'r') 
    
plt.plot(theta, expected, color = 'b') 
 
plt.legend(labels =('员工1', '员工2'),bbox_to_anchor=(1, 0.2, 0.2, 0.8), fontsize=15) 

plt.title('我是标题', fontsize = 20)
plt.tick_params(labelsize=20)

plt.show()
```

### 基础雷达图 - 加入填充

```python
import matplotlib.pyplot as plt 
import numpy as np 
    
types = ["A", "B", "C", "D", "E"] 
  
actual = [41, 57, 34, 53, 60, 41] 
expected = [40, 49, 20, 54, 65, 40] 
    
plt.figure(figsize =(9, 8)) 
plt.subplot(polar = True) 
    
theta = np.linspace(0, 2 * np.pi, len(actual)) 
    
lines, labels = plt.thetagrids(range(0, 360, int(360/len(types))), 
                                                         (types)) 

plt.plot(theta, actual) 
plt.fill(theta, actual, 'b', alpha = 0.1) 

plt.plot(theta, expected) 
plt.fill(theta, expected, 'r', alpha = 0.1) 
    
plt.legend(labels =('员工1', '员工2'),bbox_to_anchor=(1, 0.2, 0.2, 0.8), fontsize=15) 

plt.title('我是标题', fontsize = 20)
plt.tick_params(labelsize=20)

plt.show()
```
