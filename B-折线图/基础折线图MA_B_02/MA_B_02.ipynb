### 基础折线图 - 多组数据

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots(figsize=(9,8)) 
ax.plot(x, x, label='linear') 
ax.plot(x, x**2, label='quadratic') 
ax.plot(x, x**3, label='cubic') 
ax.set_xlabel('我是x轴', fontsize = 20)  
ax.set_ylabel('我是y轴', fontsize = 20)  
ax.set_title("我是标题", fontsize = 20) 
plt.tick_params(labelsize=15)
ax.legend(prop={'size':15}) 

plt.show()
```

### 基础折线图 - 金融数据可视化

```python
df = pd.read_csv("economics.csv")

x = df['date']
y1 = df['psavert']
y2 = df['unemploy']


fig, ax1 = plt.subplots(1,1,figsize=(16,9), dpi= 80)
ax1.plot(x, y1, color='tab:red')

ax2 = ax1.twinx() 
ax2.plot(x, y2, color='tab:blue')


ax1.set_xlabel('时间', fontsize=20)
ax1.tick_params(axis='x', rotation=0, labelsize=14)
ax1.set_ylabel('个人存储率', color='tab:red', fontsize=20)
ax1.tick_params(axis='y', rotation=0, labelcolor='tab:red' )
ax1.grid(alpha=.4)


ax2.set_ylabel("失业人数", color='tab:blue', fontsize=20)
ax2.tick_params(axis='y', labelcolor='tab:blue')
ax2.set_xticks(np.arange(0, len(x), 60))
ax2.set_xticklabels(x[::60], rotation=90, fontdict={'fontsize':14})
ax2.set_title("个人储蓄率和失业人数对比图", fontsize=22)
fig.tight_layout()


plt.show()
```
